```python
x=df.drop('target',axis=1)
y=df['target']
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=111)
```

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
x_train=sc.fit_transform(x_train)
x_test=sc.fit_transform(x_test)
```

```python
from sklearn.linear_model import LogisticRegression
lr=LogisticRegression()
lr.fit(x_train,y_train)

```

```python
from sklearn.metrics import accuracy_score,f1_score,confusion_matrix,classification_report
lr_pred=lr.predict(x_test)
lr_acc=accuracy_score(y_test,lr_pred)
lr_f1_score=f1_score(y_test,lr_pred)
lr_cm=confusion_matrix(y_test,lr_pred)
```

```python
print("Accuracy Score:",lr_acc)
print("F1 Score:",lr_f1_score)
print("Confusion Matrix:\n",lr_cm)
print("\n Classification Report \n",classification_report(y_test,lr_pred))
```

```python
from sklearn.tree import DecisionTreeClassifier
dt=DecisionTreeClassifier()
dt.fit(x_train,y_train)
```

```python
dt_pred=dt.predict(x_test)
dt_acc=accuracy_score(y_test,dt_pred)
dt_f1_score=f1_score(y_test,dt_pred)
dt_cm=confusion_matrix(y_test,dt_pred)

print("Accuracy Score:",dt_acc)
print("F1 Score:",dt_f1_score)
print("Confusion Matrix:\n",dt_cm)
print("\n Classification Report:\n",classification_report(y_test,dt_pred))
```

```python
from sklearn.ensemble import RandomForestClassifier
rf=RandomForestClassifier(n_estimators=100)
rf.fit(x_train,y_train)
```

```python
rf_pred=rf.predict(x_test)
rf_acc=accuracy_score(y_test,rf_pred)
rf_f1_score=f1_score(y_test,rf_pred)
rf_cm=confusion_matrix(y_test,rf_pred)

print("Accuracy Score:",rf_acc)
print("F1 Score:",rf_f1_score)
print("Confusion Matrix:\n",rf_cm)
print("\n Classification Report:\n",classification_report(y_test,rf_pred))
```
